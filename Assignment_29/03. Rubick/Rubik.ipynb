```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("rubix.png")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
#plt.imshow(img)
```

```python
R, G, B = cv.split(img)
#plt.imshow(B)
```

```python
Height, width, _ = img.shape
for i in range(Height):
    for j in range (width):
        if(G[i,j] > 127 and B[i,j] > 127 and R[i,j] < 127):
            img[i,j] = (255,0,0)
        elif(R[i,j] > 127 and B[i,j] > 127 and G[i,j] < 127):
            img[i,j] = (0,255,0)
        elif(R[i,j] > 127 and G[i,j] > 127 and B[i,j] < 127):
            img[i,j] = (0,0,255)
plt.imshow(img)
```

```python
rubik_Result = cv.cvtColor(img, cv.COLOR_RGB2BGR)
cv.imwrite("rubik_Result.jpg", rubik_Result)
```
