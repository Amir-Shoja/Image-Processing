```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import ImageFont, ImageDraw, Image  
```

```python
logo = np.ones((650, 1480, 3), dtype = "uint8") * 84
logo[0:650, 0:1480] = (87,83,82) # background
#plt.imshow(logo)
```

```python

logo[210:310,210:310] = (246,83,20) # red
logo[320:410,210:310] = (0,161,241) # blue
logo[210:310,320:420] = (124,187,0) # green
logo[320:410,320:420] = (255,187,0) # Yelow
plt.imshow(logo)
```

```python
font = ImageFont.truetype("segoeui.ttf", 190)
pil_im = Image.fromarray(logo)
draw = ImageDraw.Draw(pil_im)
draw.text((470, 175), "Microsoft", font = font)
logo = np.array(pil_im)
plt.imshow(logo)
```

```python
logo = cv.cvtColor(logo, cv.COLOR_RGB2BGR)
cv.imwrite('logo.jpg', logo)
```
